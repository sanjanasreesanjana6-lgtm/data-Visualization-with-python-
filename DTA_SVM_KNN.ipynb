```python
# 📘 Import Libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt

# 🌸 Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# 🧩 Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 🌳 Create the Decision Tree model
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, y_train)

# 🔮 Make predictions
y_pred = model.predict(X_test)

# 📊 Evaluate the model using a Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Decision Tree Confusion Matrix")
plt.show()

# ✅ Print Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# 🌳 Visualize the Decision Tree
plt.figure(figsize=(7,8))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
#this code is by using the confusion matrix
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

data = {
    'Study_Hours': [5, 15, 25, 35, 45, 10, 50, 60, 20, 30],
    'Attendance': [60, 70, 80, 90, 95, 65, 85, 100, 75, 88],
    'Result': ['Fail', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail', 'Pass', 'Pass', 'Fail', 'Pass']
}

df = pd.DataFrame(data)
X = df[['Study_Hours', 'Attendance']]
y = df['Result']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred, labels=['Pass', 'Fail'])

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].bar(['Accuracy'], [accuracy], color='skyblue')
axes[0].set_ylim(0, 1)
axes[0].set_title('Model Accuracy')
axes[0].text(0, accuracy/2, f"{accuracy*100:.1f}%", ha='center', fontsize=12)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Pass', 'Fail'])
disp.plot(ax=axes[1], cmap=plt.cm.Blues, colorbar=False)
axes[1].set_title('Confusion Matrix')

plot_tree(model,
          feature_names=['Study_Hours', 'Attendance'],
          class_names=['Fail', 'Pass'],
          filled=True, rounded=True, ax=axes[2])
axes[2].set_title('Decision Tree Structure')

plt.tight_layout()
plt.show()
```

```python
# Step 1: PCA-reduce to 2D
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

# Load and scale data
data = load_breast_cancer()
X = data.data
y = data.target
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 2: Train SVM on PCA-reduced data
model = SVC(kernel='linear')
model.fit(X_pca, y)

# Step 3: Plot decision boundary
def plot_svm_boundary(X, y, model):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.title("SVM Decision Boundary on PCA-Reduced Breast Cancer Data")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.grid(True)
    plt.show()

plot_svm_boundary(X_pca, y, model)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Class 0 points (red)
X_class0 = np.array([[1, 2], [2, 3], [3, 3], [2, 1]])

# Class 1 points (blue)
X_class1 = np.array([[6, 5], [7, 7], [8, 6], [7, 5]])

# Combine the data
X = np.vstack((X_class0, X_class1))
y = np.array([0]*len(X_class0) + [1]*len(X_class1))

# Create linear SVM model
model = SVC(kernel='linear')
model.fit(X, y)

# Predict a new point
new_point = np.array([[4, 3]])
prediction = model.predict(new_point)
print("Predicted class for point", new_point, "is:", prediction)

# Get the separating hyperplane parameters
w = model.coef_[0]
b = model.intercept_[0]

# Create a grid to plot the decision boundary
x_min, x_max = 0, 9
y_min, y_max = 0, 9
xx = np.linspace(x_min, x_max, 100)
yy = -(w[0] / w[1]) * xx - b / w[1]

# Margins
margin = 1 / np.sqrt(np.sum(w ** 2))
yy_down = yy - np.sqrt(1 + (w[0] / w[1]) ** 2) * margin
yy_up = yy + np.sqrt(1 + (w[0] / w[1]) ** 2) * margin

# Plot points
plt.scatter(X_class0[:, 0], X_class0[:, 1], color='red', label='Class 0')
plt.scatter(X_class1[:, 0], X_class1[:, 1], color='blue', label='Class 1')
plt.scatter(new_point[:, 0], new_point[:, 1],
            color='green', marker='*', s=200, label=f'New point (class {prediction[0]})')

# Plot decision boundary and margins
plt.plot(xx, yy, 'k-', label='Decision boundary')
plt.plot(xx, yy_down, 'k--', label='Margins')
plt.plot(xx, yy_up, 'k--')

# Plot support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='k', label='Support vectors')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear SVM Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```

#K-NEAREST NEIGHBOR


-->>this is also a supervised learning supports both classification and regression


-->> this is used to assign the new data point

-->>when a new data is found,the work of knn is to identify the nearest value by using the circle ,so to the nearest value the data is assigned

-->>the main addvantage of the knn is this dont need any kind of training data to learn (doesn't need any specific data)

-->>K-Nearest Neighbors is also called as a **lazy learner algorithm** because it does not learn from the **training set immediately** instead it stores the **entire dataset** and performs computations only at the time of classification.(so we can call it as a drawback)

-->>distance formulas are used(eucllidean,Manhattan,minkowski)[formula in book]

#RANDOM FOREST

-->>this is also a supervised learning supports both classification and regression(it mostly use the classification only)

-->>it uses the multiple decision trees

-->>from those it takes the majority of votes as output

For classification: it takes a majority vote.

For regression: it takes the average of predictions.

This randomness helps reduce overfitting and improves generalization.

```python
#import the necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier # Corrected import
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score
import  seaborn as sn
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
from sklearn.datasets import load_breast_cancer
import numpy as np

# Load data
data = load_breast_cancer()
X = data.data
y = data.target
df=pd.DataFrame(X,columns=data.feature_names)
print(df.shape)
df.head()
```

```python
#check if there are any null values in the data
df.isnull().sum()
```

```python
#assign the dependent and independent variables
x=data.data
y=data.target
```

```python
#spitin the data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#processing data with standard scalar
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#fitting the model random Forest Classifier
model=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
```

```python
#evaluation metrics
print('Random Forest Classifier')
conf_mat=metrics.confusion_matrix(y_test,y_predict)
print('Confusion Matrix:\n',conf_mat)
accuracy_score=metrics.accuracy_score(y_test,y_predict)
print('Accuracy score:',accuracy_score)
print('Accuracy in Percentage:',int(accuracy_score*100),'%')
print(classification_report(y_predict,y_test))
```
