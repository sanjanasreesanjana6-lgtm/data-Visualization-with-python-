#7) A list contains the average daily temperature(in degree Celsius) of a city over a particular week. Write a Python code to swap the highest and the lowest temperatures



```python
list=[48,54,16,92,63,72,95]
print("actual list:",list)
max_index=list.index(max(list))
min_index=list.index(min(list))
print("max index:",max_index)
print("min index:",min_index)
list[max_index],list[min_index]=list[min_index],list[max_index]
print("Temperatures after swapping highest and lowest:",list)
```

#9) Write a Python script to check whether a given key already exists in a dictionary. If it does, drop that key from the dictionary. If it does not, add that key and corresponding value to the dictionary

#a) Check if the key 'Employee_ID' is present in the dictionary. Hint : use a boolean variable to store this value

#b) If not present add this key to this dictionary and the corresponding value to be 4503233

#c) Check if the key 'Band' is present in the dictionary

#d) If present, drop this key from the dictionary

#e)Print the updated dictionary



```python
# Initial dictionary (example)
employee_dict = {"Name": "John Doe","Department": "IT","Band": "A2"}

# a) Check if 'Employee_ID' is present
key_exists = "Employee_ID" in employee_dict   # boolean variable

if not key_exists:
# b) If not present, add key with value 4503233
    employee_dict["Employee_ID"] = 4503233

# c) Check if 'Band' is present
if "Band" in employee_dict:
# d) If present, drop this key
    employee_dict.pop("Band")

# e) Print the updated dictionary
print("Updated Dictionary:", employee_dict)
```

#10) Having a nested list sometimes might be a bit problematic. An individual was asked to collect the names of companies in the technology sector. While creating the list, by mistake the last three companies were subsumed in a list as shown below. You are required to get rid of the nesting
#a)The list of tech companies curated by the individual
#tech_companies = ['Qualcomm','Google','Apple',['Nvidia','Cisco','Samsung']]

#b) The final list that we want

#c) correct_tech_companies = ['Qualcomm','Google','Apple','Nvidia','Cisco','Samsung']



```python
# Original list
tech_companies = ['Qualcomm','Google','Apple',['Nvidia','Cisco','Samsung']]

# Flatten the nested list
correct_tech_companies = tech_companies[:3] + tech_companies[3]

print("Correct list:", correct_tech_companies)
```
