SORT METHOD:: sort and sorted is different from each other

sort is a **method** and sorted is a **function** ---sort will perminently change the list, but whereas sorted will not perminently change the list

and the sort list should contain the same data type obviously

-->> by defalut it arranges in ascending order but if u want in decending order then we have to use the REVERSE=TRUE





```python
sanjana=[1,9,2,3,4,5,6,7,8,9]
sanjana.sort()
sanjana
```

```python
sanjana=['sun','mon','tue','wed','thu','fri','sat']
sanjana.sort()
sanjana
# it just arranged them according to the alphabitical order
```

```python
sanjana.sort(reverse=True)
sanjana
```

```python
vaishu=['true','false','wrong','correct']
vaishu.sort()
print(vaishu)
vaishu.sort(reverse=False)
print(vaishu)
vaishu.sort(reverse=True)
print(vaishu)
```

ZIP FUNCTION:: is used to map the 2 lists,it is not same as nested list

-->> it maps the elements based on the index values 0 with 0,1 with 1,.....

```python
classes=['cn','daa','da','irs','devops',1]
days=['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
zip(classes,days)
```

```python
waa=list(zip(classes,days))
waa
```
