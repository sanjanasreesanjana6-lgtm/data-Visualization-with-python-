**CHART/GRAPH**this is used to understand the concept easily

-->> this is all about converting the raw data into a structured format

-->> There are many kind of charts around 15 graphs


-->>we have to decide which one is suitable to use

**1 BAR GRAPH**:categorical,quantative

-->> the data will be vertically or horizontally

-->>it works for the minimum categories only we should avoid it for the more categories

**2 LINE CHART**:continuous, time-series

-->>are used when trends to the time,x-axis represents the time and y-axis represents the value

-->>use this when there is the logical order and relationship between the datapoints only

**3 DONUT GRAPH**::categorical,proportional

-->> used to show the proportions of the data(ratio,percentage).it is in the circular shape

-->>avoid when they're many categories and comparing among the groups

**4 SCATTER PLOT**::contnous,bivairate

-->> use when two varibles are needed ,it shows the two varible plane one on x axis and other on y axis

-->>not accept the muliple values and relationship with variables are not allowed

**5 AREA CHART**::continuous,time series

-->> it is same as the line chart but it covers the area,it emphasizes the magnitude and volume

-->>avoid when multple area involve and overlaping is present

**6 BUBBLE CHART**::continuous,multivariate


-->>ue when we need to show the relationship between the 3 varibles,shows the range according to the size

-->>avoid when the size of the bubble i not meaningful and multiple values also

**7 HISTO GRAM**::continuous,univariate(single)

-->>neariy looks same as bar graph ,but it is not abar graph because it is continuous and no space between the bars,the data is also divided into equal intervals

-->>avoid using the histo grams when the data is categorical and comparing across the data groups

**8 HEAT MAP**::continuous,multivariate

-->>this is used to represent the difference between the two variables by coloring them

-->>one on x axis and other on y axis,dark colour indicates the higher value and vice versa

-->> dont use the heat map when the comparison between the variable is irrelevant,or comparing mutiple categories


**9 TREE MAP**::categorical,hierarchical


-->>it similar to bubble chart but the representation wi be in the circle form ,but here it is in the form of rectangle and square

-->> size indicates the value

-->> avoid when there are too many categories or data is not hierarchical

**10 RADAR CHART**::continuous,multivariate

-->> it aslo connects the dots to form a dimensional shape(nearly circles)

-->>avoid when they are only a few dimensions to compare,comparing across the data groups

**11 box plot/whisker plot**::continuous,univariate

-->>use to display the distribution of data and detect the outlier's

-->>box plot use the rectangles to represent the interquartile range(IQR),the median is present in the box for variation

-->> avoid using the box plot when data is categorical or showing trends over time

**12 stacked barchart**::categorical,quantitative

-->>slightly different to the bar graph,no axises are present

-->>represented by rectangles,used to show the comparison between the columns , the length of each segment repreent the value

-->>avoid when there are too many categories or the data is continuous

**13 GANTT CHART**::time-based,project management

-->>use to visualize the project schedules,tasks and milestones

-->>uses horizontal bars the size of bar matters

-->>avoid when the data is not time based or showing the relationship between the variables

**14 WATERFALL CHART**quantitative,sequential

-->>use to visualize the cumulative of sequence data,such as financial or inventory changes

-->>uses vertical bars,postive and negative values are differentiated by color's

-->>avoid when there is no logical order or relationship between the data points

**15 FUNNEL CHART**categorical,process states

-->>use to visualize the stages of a process,such as sales,conversions,or customer journeys

-->>represented by using the stages,each stage is a process

-->>avoid when there is no straightforward process or data is continuous
