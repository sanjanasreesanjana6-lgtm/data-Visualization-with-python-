```python
import matplotlib.pyplot as plt

# Example data (replace with your actual data)
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
costs = [2500, 1500, 3000, 100000, 4999, 98444, 23000]
items = ['dress', 'top', 'jeans', 'saree', 'shirt', 'hat', 'coat']

# Line Chart
plt.figure(figsize=(8, 4)) # Optional: Adjust figure size
plt.plot(years, costs, marker='o') # marker='o' adds points to the line
plt.xlabel('Year')
plt.ylabel('Cost')
plt.title('Cost Over Time (Line Chart)')
plt.grid(True) # Optional: Add a grid
plt.show()

# Donut Chart
plt.figure(figsize=(6, 6)) # Optional: Adjust figure size
# Create a pie chart
plt.pie(costs, labels=items, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'white'})

# Draw a circle in the center to make it a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Costs (Donut Chart)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
