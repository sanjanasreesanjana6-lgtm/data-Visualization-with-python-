UNION OPERATOR::union() returns the union of the two sets

-->>union is denoted by |(or)

INTERSECTION OPERATOR :: intersection() returns the same elements

-->> denoted by &(and)

DIFFERENCE OPERATOR :: difference() returns the non same elements

-->> denoted by -(minus)

NOTE::: a-b returns the non same elements of a
        b-a returns the non same elements of b

SYMMETRIC_DIFFERENCE OPERATOR:: symmetric_difference() returns the both the elements from the sets but not the same elements (excluding the intersection elements)

-->> denoted by ^(caret)

```python
a={1,2,34,6,7,8}
c={1,2,3,5,4,8,9,7}
a.union(c)
```

```python
a|c
```

```python
a&c
```

```python
a={1,2,34,6,7,8}
c={1,2,3,5,4,8,9,7}
a.intersection(c)
```

```python
a.difference(c)
```

```python
c.difference(a)
```

```python
a-c
```

```python
c-a
```

```python
a={1,2,34,6,7,8}
c={1,2,3,5,4,8,9,7}
a.symmetric_difference(c)
```

```python
c.symmetric_difference(a)
```

```python
a^c
```

```python
c^a
```

```python
# recreate the dictionary
sanj=['fish',2,3,4,5,6]
don=['san','pan','man','god','fan','tom']
san_don=zip(sanj,don)
dict_san_don=dict(san_don)

dict_san_don.pop('fish')
print(dict_san_don)

dict_san_don.pop(2)
print(dict_san_don)

```
