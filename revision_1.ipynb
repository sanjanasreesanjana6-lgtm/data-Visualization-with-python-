```python
#plotly is also a library used in Visualization
#ols[Ordinary Least Squares]=it's one of the most widely used techniques in linear regression.
#Minimize the squared distance between each point and the line â€” those distances are called residuals.
#it is used to reduce the errors
#keep the mouse at the each point u will observe the information of each data
#so by this we can minimize the data
import plotly.express as px
import plotly.graph_objects as go
figure=px.scatter(data_frame=sales,x="Quantity",y="TotalPrice",size="Quantity",trendline="ols")
figure.show()
```

```python
#just observe this
import plotly.express as px
import plotly.graph_objects as go
figure=px.scatter(data_frame=sales,x="Cashier",y="TotalPrice")
figure.show()
```

```python
#this is used to the correlation between them
#correlation works only for the numerical values
feature=["Quantity","TotalPrice","UnitPrice"]
correlation=sales[feature].corr()
print(correlation)
```

```python
x=sales[feature]
y=sales.TotalPrice
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
display(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_prediction=model.predict(x_test)
```

```python
print(y_prediction)
print(y_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,mean_squared_log_error
```

```python
print("r-square value:",r2_score(y_prediction,y_test))
```

```python
#if the error value is less then the acurracy is good
print("mean abosuate error:",mean_absolute_error(y_prediction,y_test))
```

```python
print("mean squared error:",mean_squared_error(y_prediction,y_test))
```

```python
print("mean squared log error:",mean_squared_log_error(y_prediction,y_test))
```
